```python
import torch
import matplotlib.pyplot as plt
```

```python
alphas = torch.linspace(1e-10, 1.0, 4)
colors = ["r", "g", "b", "y", "k"]

x = torch.arange(1, 10)[None, :]
y = torch.pow(alphas[:, None], x)
y_norm = torch.nn.functional.normalize(y, p=1, dim=-1)
y_soft = torch.softmax(y, dim=-1)
y_norm_soft = torch.softmax(y_norm, dim=-1)
for i in range(y.shape[0]):
    plt.plot(x[0], y_norm[i], label=f"alpha={alphas[i].item():.2f}", color=colors[i])
    plt.plot(x[0], y_soft[i], label=f"alpha={alphas[i].item():.2f}(softmax)", linestyle="--", color=colors[i])
    plt.plot(x[0], y_norm_soft[i], label=f"alpha={alphas[i].item():.2f}(norm+softmax)", linestyle=":", color=colors[i])
plt.legend()
plt.show()
```
